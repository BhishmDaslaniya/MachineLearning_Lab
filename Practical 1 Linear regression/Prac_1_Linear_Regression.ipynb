**Aim:** Build a regression model and predict the per capita income fo canadian citizens in year 2020.

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn import linear_model
```

```python
df = pd.read_csv('canada_per_capita_income.csv')
df
```

### Graph of given data

```python
%matplotlib inline
plt.xlabel('year')
plt.ylabel('per capita income (US$)')
plt.scatter(df['year'],df['per capita income (US$)'],color='red',marker='+')
```

```python
year = df[['year']]
year
```

```python
income = df['per capita income (US$)']
income
```

```python
reg = linear_model.LinearRegression()
reg.fit(year,income)
```

```python
reg.predict([[2020]])
```

### Answer = 41288.69409442

```python
reg.coef_ # m
```

```python
reg.intercept_ # b
```

```python
# Verify by putting values in linear equation
# Y = mX + b
828.46507522 * 2020 + -1632210.7578554575
```

```python
predicted_income = reg.predict(year)
predicted_income
```

```python
predicted_df = pd.DataFrame(year)
predicted_df.head(10)
```

```python
predicted_df['predicted income'] = predicted_income
predicted_df.head(10)
```

### Graph of actual data and predicted data by linear regression model

```python
%matplotlib inline
plt.xlabel('year', fontsize=20)
plt.ylabel('per capita income (US$)', fontsize=20)
plt.scatter(df['year'],df['per capita income (US$)'],color='red',marker='+')
plt.plot(predicted_df['year'],predicted_df['predicted income'],color='blue')
```
